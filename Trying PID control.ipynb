```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
game_mode="1D"
target_pos=0
present_pos=9
present_approach_speed=3 #UnitsPerSecond
prop_factor=0.5
accepted_error=0.5
list_of_pos=np.array([present_pos])
time_tested_for=100
accelaration_deacc_limit=0.33 #0.33 unit vel per second

def calc_new_pos(present_pos,present_approach_speed):
    new_pos=present_pos+present_approach_speed
    return new_pos

def calc_new_approach_speed(prop_factor,present_approach_speed,present_pos,target_pos):
    new_approach_speed=prop_factor*(target_pos-present_pos) 
    return new_approach_speed

for i in range(1,100):
    present_pos=calc_new_pos(present_pos,present_approach_speed)
    list_of_pos=np.append(list_of_pos,present_pos)
    if np.abs(present_pos-target_pos)< accepted_error:
        break
    present_approach_speed=calc_new_approach_speed(prop_factor,present_approach_speed,present_pos,target_pos)

fig, ax=plt.subplots()
ax.plot(list_of_pos)
```
